```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def given_function(x):
    return (x+3)**2
def derivative(x):
    return 2*(x+3)
```

```python
def gradient_descent(function,start,learn_rate,n_iter=100,tolerance=0.1):
    gradient=derivative
    function=given_function
    points=[start]
    iters=0

    while iters<n_iter:
        prev_x=start
        start=start-learn_rate*gradient(prev_x)
        iters+=1
        points.append(start)
    print("Local Minimum occurs at:",start)

    x_=np.linspace(-7,5,10)
    y=function(x_)

    fig=plt.figure(figsize=(10,10))
    plt.plot(x_,y,"g")
    plt.plot(points,function(np.array(points)),"-o")
    plt.show()
        
        
        
```

```python
gradient_descent(function=given_function,start=2.0,learn_rate=0.2,n_iter=10)
```
